```python
import pptk
import laspy
import numpy as np

lidar_point_cloud = laspy.read("lidar.laz")
```

```python
# Getting the number of points in the dataset and available dimensions
number_of_points = lidar_point_cloud.header.point_count
dimensions = list(lidar_point_cloud.point_format.dimension_names)

print("Number of points in dataset: ", number_of_points)
print("Available dimensions in dataset: ", dimensions)
```

```python
# Create Numpy array from the X, Y, Z coordinates
# Note: Use lowercase .x, .y, .z for scaled coordinates in laspy
coords = np.vstack((lidar_point_cloud.x, lidar_point_cloud.y, lidar_point_cloud.z)).transpose()

# Use 'intensity' since your file doesn't have red, green, or blue
# We repeat it three times to create a grayscale effect (R=G=B)
colors = np.vstack((lidar_point_cloud.intensity, lidar_point_cloud.intensity, lidar_point_cloud.intensity)).transpose()
```

```python
# initialize PPTK viewer
viewer = pptk.viewer(coords, colors)
```

```python
# Create the coordinate array (X, Y, Z)
coords = np.vstack((lidar_point_cloud.x, lidar_point_cloud.y, lidar_point_cloud.z)).transpose()

# Initialize the viewer using intensity for the color values
viewer = pptk.viewer(coords, lidar_point_cloud.intensity)

# Apply a color scheme (like 'jet') to make the intensity visible
viewer.color_map('jet')
```
